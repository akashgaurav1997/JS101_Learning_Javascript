#### ****1. Which Neighbourhood Group has more listing?****

```python
#Using a countplot to plot the neighbourhood group with number of listing:
plt.figure(figsize=(10,6))
sns.countplot(data = df_air, x = 'neighbourhood_group')
plt.title('Neighbourhood With Most Hotel Listing')
plt.xlabel('States Name')
plt.show()
```

**The Countplot will count all the listings in all the Neighbourhood and compare all the state clearly using the barchart.**

**Observation:**
 * Manhattan has more hotel listing comparing to other states.
 * Staten Islan has least hotel listing comparing to other states.

#### ****1.1 Which Neighbourhood has more listing?****

```python
#We are trying to draw a countplot for top 20 neighbourhood with more number of hotels.
plt.figure(figsize=(20,6))
sns.countplot(x = df_air['neighbourhood'], order = df_air['neighbourhood'].value_counts().iloc[:20].index)
plt.title('Hotel Listing In Each Neighbourhood', fontsize = 14)
plt.xlabel('Neighbourhood Names', fontsize = 12)
plt.ylabel('Count Of Hotels', fontsize = 12)
plt.xticks(rotation = 90, fontsize = 8)
plt.show()
```

#### ****2. Which Room Type has more listing?****

```python
#We are using countplot to show the count of all room types:
plt.figure(figsize=(10,5))
sns.countplot(data = df_air, x = 'room_type')
plt.title('Room Types With More Listing')
plt.xlabel('Room Types')
plt.show()
```

 **Observation:**
  * We can observe that Entire home/ apartment has more listing on AirBnb.
  * We can observe that Shared Room has least listing on AirBnb.

#### ****3. Which Room Types has highest average price?****

```python
#We are grouping room_type with price and finding the mean of the price and saving in a variable to be used to plot distribution barchart:
df1 = df_air.groupby('room_type')['price'].mean().reset_index()

#We are  ploting a barchart to show the room_type with highest average price:
plt.figure(figsize=(10,5))
sns.barplot(data = df1, x = 'room_type', y = 'price')
plt.title('Average Room Price')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```

**Observation:**
 * We can observe that Entire home/apt has highest average price, followed by Private Room.
 * We can observe that Shared Room has least average price.

#### ****4. Which Room Type is costliest in each Neighbourhood?****

```python
#We will group neighbourhood and roomtype and find the mean of the price.
df_neighbourhood_price = df_air.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()

#we will draw a bar plot to show the distribution of price for each roomtype and neighbourhood.
plt.figure(figsize=(10,5))
sns.barplot(data = df_neighbourhood_price, x = 'neighbourhood_group', y = 'price', hue = 'room_type')
plt.title('Average Room Price In Each Neighbourhood')
plt.xlabel('States')
plt.ylabel('Price')
plt.show()
```

**Observation:**
 * We can observe that Entire home/apt is costliest in each neighbourhood, followed by Private Rooms.
 * We can observe that Shared Rooms are most affordable in each neighbourhood.

#### ****6. What Is The Average Minimum Nights Stay For Each Room Type?****

```python
#We tried to find the mean of minimum nights.
df18 = df_air.groupby('room_type', as_index = False).agg({"minimum_nights":"mean"}).sort_values("minimum_nights", ascending = False)
df18
```

```python
#We have tried to draw a barchart to show the average minimum nights with each room type.
plt.figure(figsize=(10,6))
sns.barplot(data = df18, x = 'room_type', y = 'minimum_nights')
plt.title('Average Minimum Nights Stayed', fontsize = 12)
plt.xlabel('Room Types', fontsize = 12)
plt.ylabel('Average Of Minimum Nights Stayed')
plt.show()
```

**Observations:**
 * We can observe that Entire home/apt has maximum stay, followed by shared room .
 * We can observe that Private room has least stay and that's not a surprise as they are more expensive for indivduals.

#### ****7. What Is The Average Rent Paid By Each Person For Each Room Type?****

```python
#We tried to find the mean of price and mean of guests visited by each room type.
df19 = df_air.groupby('room_type', as_index = False).agg({"price":"mean", 'calculated_host_listings_count':'mean'}).sort_values("room_type", ascending = False)
df19

```

```python
#We tried to find the average of rent paid by each person.
avg_price_per_person = df19['price']/df19['calculated_host_listings_count']
avg_price_per_person
```

```python
#We tried to draw a barplot to show the average rent paid by each guest for each room type.
plt.figure(figsize=(10,6))
sns.barplot(x = df19['room_type'], y = avg_price_per_person)
plt.title('Average Rent Paid By Each Guest', fontsize = 14)
plt.xlabel('Room Types', fontsize = 12)
plt.ylabel('Average Rent Paid', fontsize = 12)
plt.show()
```

**Observations:**
 * We can observe that Private rooms are more expensive for individual.
 * We can observe that Shared rooms are quite affordable for individual.

#### ****11.In Which Year Room Price Was High?****
